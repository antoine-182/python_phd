```python
# -*- coding: utf-8 -*-

import netCDF4 as nc4
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator
```

```python
dirt = "/gpfswork/rech/omr/ukd13rj/nemo/penalnemov4.0.1/tests/penaloverflow/exp_cfl/"
#dirm = "/gpfswork/rech/bho/ukd13rj/nemo/dev_r12527_Gurvan_ShallowWater/cfgs/article_CORNER/MESHMASK/"

#workdir="moy_phimin01_smo2"
#workdir="moy_phimin0001_smo6"
#workdir="ref_zps_AimpT"
#workdir="ref_zco_AimpT"
dirt = "/gpfswork/rech/omr/ukd13rj/nemo/penalnemov4.0.1/tests/penaloverflow/exp_up1/"
         
flagbpv=True

pdt = dirt + workdir + "/OVF_grid_T.nc"
pdu = dirt + workdir + "/OVF_grid_U.nc"

save = 1 ; psave = "ratios/rphi_%s.png" % (workdir)
dpi = 600 
```

```python
""" Load data
"""
tb=0
dt = nc4.Dataset(pdt)

if flagbpv : 
    rpot    = dt.variables['rpot'   ][tb,:,1,:]
    rlipst  = dt.variables['rlipst' ][tb,:,1,:]
    rphiu_t = dt.variables['rphiu_t'][tb,:,1,:]
    rphiw_t = dt.variables['rphiw_t'][tb,:,1,:]
#
dt.close()

du = nc4.Dataset(pdu)
if flagbpv :
    rpou    = du.variables['rpou'   ][tb,:,1,:]
    rlipsu  = du.variables['rlipsu' ][tb,:,1,:] 
    rphiu_u = du.variables['rphiu_u'][tb,:,1,:]
    rphiw_u = du.variables['rphiw_u'][tb,:,1,:]
#
du.close()
print("data loaded")

""" Figure
""" 

palette = plt.get_cmap('RdBu_r',10)
optpcolor = {"cmap" : palette}
optptoce = {"cmap" : palette, "vmin" : 10, "vmax" : 20}

NI = 2 ; NJ = 4 # nb of rows (NI) and cols (NJ)
fig, ax = plt.subplots(NI,NJ, figsize=(NJ*6,NI*4), dpi = dpi, squeeze=False)
# settled per rows (i=0) j=0 j=1 j=2
#                  (i=1) j=0 j=1 j=2
im = [[],[],[]]

""" 1st row : toce lipsu rlipsu lipst rlipst
"""
im[0].append( ax[0][0].imshow(rpot     , **optpcolor)  )  ; ax[0][0].set_title('rpot - %s' % (workdir))
im[0].append( ax[0][1].imshow(rlipst   , **optpcolor)  )  ; ax[0][1].set_title(' rlipst max %.1f' % ( np.max( rlipst)) )
im[0].append( ax[0][2].imshow(rphiu_t  , **optpcolor)  )  ; ax[0][2].set_title('rphiu_t max %.1f' % ( np.max(rphiu_t)) )
im[0].append( ax[0][3].imshow(rphiw_t  , **optpcolor)  )  ; ax[0][3].set_title('rphiw_t max %.1f' % ( np.max(rphiw_t)) )
""" 2nd row : cflu cfluu rphiu_u cfluw rphiw_u
"""
im[1].append( ax[1][0].imshow(rpou     , **optpcolor) )  ; ax[1][0].set_title('rpou')
im[1].append( ax[1][1].imshow(rlipsu   , **optpcolor) )  ; ax[1][1].set_title(' rlipsu max %.1f' % ( np.max( rlipsu)) )
im[1].append( ax[1][2].imshow(rphiu_u  , **optpcolor) )  ; ax[1][2].set_title('rphiu_u max %.1f' % ( np.max(rphiu_u)) )
im[1].append( ax[1][3].imshow(rphiw_u  , **optpcolor) )  ; ax[1][3].set_title('rphiw_u max %.1f' % ( np.max(rphiw_u)) )

""" pretty
"""
#fig.colorbar(cf, ax=ax[i:(i+1)][j:(j+1)], shrink=0.8, location='bottom'),
#                 aspect = 60, fraction=0.05,  extend = 'both', pad = 0.02)
for i in range(NI):
    for j in range(NJ):
        ax[i][j].set_xticklabels([]) ; ax[i][j].set_yticklabels([])
        ax[i][j].set_xticks([])        ; ax[i][j].set_yticks([]) 
        ax[i][j].set_aspect(aspect='equal')
        fig.colorbar(im[i][j],ax=ax[i:(i+1),j:(j+1)], shrink=0.8, location='bottom',
                    pad = 0.02)
        
fig.patch.set_facecolor('white')
#plt.subplots_adjust(wspace=0.04, hspace=0.08)
if save :
    print("\nsaving : %s" % psave)
    fig.savefig(psave, dpi = dpi)
    plt.close()
    print("figure closed")
plt.show()
```
