```python
# -*- coding: utf-8 -*-

import netCDF4 as nc4
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator, MultipleLocator
```

```python
dirt = "/gpfswork/rech/bho/ukd13rj/nemo/dev_r12527_Gurvan_ShallowWater/cfgs/article_CORNER/"
dirm = "/gpfswork/rech/bho/ukd13rj/nemo/dev_r12527_Gurvan_ShallowWater/cfgs/article_CORNER/MESHMASK/"

pdF = "CORNER_1d_00010101_00010215_grid_F.nc"
pmm = "mesh_mask_1250.nc"

ListC = [[0 , "EXP_een/output.init.nc"],
         [3, "EXP_een/"+pdF],
         [20, "EXP_een/"+pdF],
         [-1, "EXP_een/"+pdF]]

Listdt = [ListC]
```

```python
########################################################

mm = nc4.Dataset(dirm+pmm)
tmask = mm.variables['tmask'][0,0]
nY,nX = np.shape(tmask)
fmask_vorlatT = np.zeros((nY,nX)) ; fmask_vorlatF = np.zeros((nY,nX))
for i in range(nX-1):
    for j in range(nY-1):
        fmask_vorlatT[j,i] = tmask[j+1,i] * tmask[j+1,i+1] \
                           * tmask[j  ,i] * tmask[j  ,i+1]
        fmask_vorlatF[j,i] = tmask[j+1,i] * tmask[j+1,i+1] \
                           * tmask[j  ,i] * tmask[j  ,i+1]
        # exception vorlat
        if ((tmask[j+1,i] + tmask[j+1,i+1] \
            + tmask[j  ,i] + tmask[j  ,i+1])) == 3 :
            fmask_vorlatT[j,i] = 1.

glamt = mm.variables['glamt'][0]/1E3 ; gphit = mm.variables['gphit'][0]/1E3 # corner of cells
glamf = mm.variables['glamf'][0]/1E3 ; gphif = mm.variables['gphif'][0]/1E3 # F points
```

```python
vmax = 0.6e-5 ; tskip=1
Ncolor=9
dv=2*vmax/Ncolor
dx = 1.25 # km

########################################################

palette = plt.get_cmap('RdBu_r',Ncolor)
#cticks = np.arange(-vmax+dv,vmax,2*dv)
cticks = np.concatenate((np.arange(-vmax,0,dv)[::2],np.flip(-np.arange(-vmax,0,dv)[::2])))
print(cticks)
optpcolor = {"vmin":-vmax, "vmax":vmax, "cmap" : palette, "alpha" : 1.0}
optcontour = {"vmin":-vmax,"vmax":vmax, "levels":cticks, "linewidths" :0.8, "colors":('k',),"linestyles" : "solid"}


save = 1 ; psave = "figure13-stopmotion.png" ; dpi = 800

NI = 1 ; NJ = 4 # nb of rows (NI) and cols (NJ)
fig, ax = plt.subplots(NI,NJ, figsize=(NJ*3,NI*3), dpi = dpi, squeeze=False)
# settled per rows (i=0) j=0 j=1 j=2
#                  (i=1) j=0 j=1 j=2

""" *************************** DATA
"""
########################################################
#init
N=len(Listdt)
data=np.zeros((nY,nX))

for i in range(NI):
    for j in range(NJ):
        frame,dirname = Listdt[i][j]
        print("[%d,%d] : %s" % (i,j,dirname))
        if dirname == 'x':
            ax[i][j].axis('off')
        else :
            dtf = nc4.Dataset(dirt+dirname)
            if frame == 0:
                data[:,:] = dtf.variables['relvor'][0,0,:,:]   # zeta
            else :
                data[:,:] = dtf.variables['relvor'][frame,:,:]
            data = np.ma.masked_where(fmask_vorlatF==0,data)
            dtf.close()
            # pcolormesh
            cf = ax[i][j].pcolormesh(glamt, gphit,data ,**optpcolor)
            # contour
            ct= ax[i][j].contour(glamf, gphif, data, **optcontour)

""" *************************** PRETTY
"""
for i in range(NI):
    for j in range(NJ):
        vorlat,dirname = Listdt[i][j]
        if dirname == 'x':
            continue
        # print(tickszer)
        ax[i][j].set_xlim(-8,508)
        ax[i][j].set_ylim(-8,508)
        ax[i][j].set_xticklabels([]) ; ax[i][j].set_yticklabels([])
        #
        ax[i][j].set_xticks([0,200,400])
        ax[i][j].set_yticks([0,200,400])
        #
        # if (i==NI-1):
        #     ax[i][j].set_xticklabels(["0","200","400"])
        #     ax[i][j].set_xlabel("X (km)")
        # #
        # if (j==NJ-1):
        #     ax[i][j].set_yticklabels(["0","200","400"])
        #     ax[i][j].set_ylabel("Y (km)")
        #     ax[i][j].yaxis.set_label_position("right")
        #
        ax[i][j].patch.set_color('0.8')
        ax[i][j].xaxis.set_minor_locator(MultipleLocator(50))
        ax[i][j].yaxis.set_minor_locator(MultipleLocator(50))
        ax[i][j].tick_params(axis = "both", which = 'both', width=1., labelsize = 5, pad = 3, \
                             bottom = True, top = True, left = True, right = True)
        ax[i][j].tick_params(which='minor',length = 3)
        ax[i][j].tick_params(which='major',length = 5)
        ax[i][j].set_aspect(aspect='equal') # data coordinate 'equal'

plt.subplots_adjust(wspace=0.04, hspace=0.08)

#cbar=plt.colorbar(cf, ax=ax[:], shrink=0.8, location = "bottom",
#                  aspect = 40, fraction=0.05,  extend = 'both', pad = 0.05) #
#cbar.set_ticks(cticks)
#pticks = np.copy(cticks)
#pticks[2:4]=0.
#cbar.set_ticklabels(["%1.e" % s if s !=0. else " " for s in pticks ])
#cbar.ax.tick_params(size=0)
#for _ in cticks :
#    cbar.ax.axvline(x=_, c='black', lw = 1.0)
#cbar.set_label(r"Relative vorticity $\zeta/h \ [\mathrm{m}^{-1}\mathrm{s}^{-1}]$")

""" *************************** SAVING
"""

if save :
    print("\nsaving : %s" % psave)
    fig.savefig(psave, dpi = dpi)
    plt.close()
    print("figure closed")
plt.show()
```
